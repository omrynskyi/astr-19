```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,1,100)

def sin(x):
    return np.sin(x)
def cos(x):
    return np.cos(x)

```

```python
siny=[]
for i in x: siny.append(sin(i))

cosy=[]
for i in x: cosy.append(cos(i))
    
fig, (axs1,axs2) = plt.subplots(1,2)

axs1.set_title("Sin(x)")
axs2.set_title("Cos(x)")
axs1.plot(x,siny)
axs2.plot(x,cosy)
plt.savefig("Session8.pdf", format="pdf")
```
